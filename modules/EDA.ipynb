```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from pathlib import Path
```

```python
DATA_DIR = Path('..', 'assets')
```

```python
marketing_path = DATA_DIR / 'GA_FinalProject.csv'
marketing = pd.read_csv(marketing_path)
```

```python
marketing.head()
```

```python
marketing.fillna(0)
```

```python
marketing.describe()
```

```python
import seaborn as sns

marketing.plot(kind='scatter', x='Total Sends', y='Positive Event')
marketing.plot(kind='scatter', x='Total Opens', y='Positive Event')
marketing.plot(kind='scatter', x='Total Clickthroughs', y='Positive Event')
marketing.plot(kind='scatter', x='Total Page Views', y='Positive Event')
marketing.plot(kind='scatter', x='Total Visits', y='Positive Event')
marketing.plot(kind='scatter', x='Average Page Time (seconds)', y='Positive Event')
marketing.plot(kind='scatter', x='Total Page Time (seconds)', y='Positive Event')
marketing.plot(kind='scatter', x='Total Form Submissions', y='Positive Event')
```

```python
ax = sns.heatmap(marketing.corr(), vmin=-1, vmax=1, cmap=sns.diverging_palette(h_neg=220, h_pos=10, n=21))
```
